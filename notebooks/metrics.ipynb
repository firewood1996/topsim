```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]


NPOINTS = 100
np.random.seed(NPOINTS)
observation_ingest = None
ya1 = [60 for x in range(20)]
yamid = [40 for x in range(10)]
ya2 = [60 for x in range(20)]
ya3 = [60 for x in range(40)]
ya4 = [20 for x in range(10)]
observation_ingest = np.concatenate((ya1,yamid,ya2,ya3,ya4), axis=None)
cumulative_buffer_storage = -np.cumsum(np.concatenate((ya1,yamid, ya2,ya3,ya4), axis=None))

yb1 = np.zeros(10)
# yb1 = [None for x in range(10)]
yb = np.random.randint(40,41,90)
# yb2 = [40 for x in range(90)]

ybsum = np.concatenate((yb1,yb),axis=None)
# yb = [50 for x in range(10)]
tmp = (cumulative_buffer_storage+np.cumsum(ybsum))
x2 = [i for i in range(10,100)]

ydiff = np.diff(yb)
yc = ((yb + np.append(ydiff,0))/-tmp[10:])
x = [i for i in range(0,100)]

fig, ax = plt.subplots()

ax.bar(x, observation_ingest, color='orange')
ax.bar(x,tmp, color='blue')
# ax.plot(x,yb,color='red')


minya, maxya = ax.get_ylim()

ax.set_ylim(minya, -minya)
ax.set_yscale('symlog')
ax2 = ax.twinx()
ax2.plot(x2, yc, color='red')
(miny, maxy) = ax2.get_ylim()
# align_yaxis(ax, 0, ax2, 0)
ax2.set_ylim(-maxy, maxy)
# n, bins, patches = ax.hist(x, 100, density=False)
plt.show()
```
